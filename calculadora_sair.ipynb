Instruções do projeto
Faça uma função calculadora que os números e as operações serão feitas pelo usuário. O código deve ficar rodando infinitamente até que o usuário escolha a opção de sair. No início, o programa mostrará a seguinte lista de operações:

1: Soma
2: Subtração
3: Multiplicação
4: Divisão
0: Sair

Digite o número para a operação correspondente e caso o usuário introduza qualquer outro, o sistema deve mostrar a mensagem “Essa opção não existe” e voltar ao menu de opções.

Após a seleção, o sistema deve pedir para o usuário inserir o primeiro e segundo valor, um de cada. Depois precisa executar a operação e mostrar o resultado na tela. Quando o usuário escolher a opção “Sair”, o sistema irá parar.

É necessário que o sistema mostre as opções sempre que finalizar uma operação e mostrar o resultado. 

Trabalhe esse código em seu IDE, suba ele para sua conta no GitHub e compartilhe o link desse projeto no campo ao lado para que outros desenvolvedores possam analisá-lo.

```python
def soma(x, y):
    return x + y

def subtracao(x, y):
    return x - y

def multiplicacao(x, y):
    return x * y

def divisao(x, y):
    if y != 0:
        return x / y
    else:
        return "Erro: divisão por zero"

def calculadora():
    while True:
        print("\nBem Vindo a Calculadora")
        print("\nEscolha a operação:")
        print("1: Soma")
        print("2: Subtração")
        print("3: Multiplicação")
        print("4: Divisão")
        print("0: Sair")

        opcao = input("Digite o número para a operação desejada: ")

        if opcao == '0':
            print("\n      Encerrando Calculadora.")
            print("\n       Obrigado Por Utilizar")
            break
        elif opcao in {'1', '2', '3', '4'}:
            try:
                numero1 = float(input("Digite o primeiro valor: "))
                numero2 = float(input("Digite o segundo valor: "))

                if opcao == '1':
                    print("Resultado:", soma(numero1, numero2))
                elif opcao == '2':
                    print("Resultado:", subtracao(numero1, numero2))
                elif opcao == '3':
                    print("Resultado:", multiplicacao(numero1, numero2))
                elif opcao == '4':
                    print("Resultado:", divisao(numero1, numero2))
            except ValueError:
                print("Entrada inválida. Certifique-se de digitar números válidos.")
        else:
            print("Essa opção não existe. Por favor, escolha uma opção válida.")

if __name__ == "__main__":
    calculadora()
```

Otimizado utilizar auxilio internet

A otimização feita utilizando um dicionário correr as operações diretamente para as funções correspondentes, evitando a necessidade de uma estrutura condicional extensa.

```python
def soma(x, y):
    return x + y

def subtracao(x, y):
    return x - y

def multiplicacao(x, y):
    return x * y

def divisao(x, y):
    return x / y if y != 0 else "Erro: divisão por zero"

def calculadora():
    operacoes = {
        '1': soma,
        '2': subtracao,
        '3': multiplicacao,
        '4': divisao
    }

    while True:
        print("\nBem Vindo a Calculadora")
        print("\nEscolha a operação:")
        print("1: Soma")
        print("2: Subtração")
        print("3: Multiplicação")
        print("4: Divisão")
        print("0: Sair")

        opcao = input("Digite o número para a operação desejada: ")

        if opcao == '0':
            print("Encerrando a calculadora.")
            break
        elif opcao in operacoes:
            try:
                numero1 = float(input("Digite o primeiro valor: "))
                numero2 = float(input("Digite o segundo valor: "))
                print("Resultado:", operacoes[opcao](numero1, numero2))
            except ValueError:
                print("Entrada inválida. Certifique-se de digitar números válidos.")
        else:
            print("Essa opção não existe. Por favor, escolha uma opção válida.")

if __name__ == "__main__":
    calculadora()
```
